```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('/content/train.csv')
df.head()
```

```python
df.isnull().sum()/len(df)*100
```

```python
df.drop(['Customer Id', 'Artist Name'], axis = 1, inplace=True)
```

```python
df['Scheduled Date']=pd.to_datetime(df['Scheduled Date'],format='%m/%d/%y')
df['Sday']=df['Scheduled Date'].dt.day
df['Smonth']=df['Scheduled Date'].dt.month
df['Syear']=df['Scheduled Date'].dt.year
```

```python
df['Delivery Date']=pd.to_datetime(df['Delivery Date'],format='%m/%d/%y')
df['Dday']=df['Delivery Date'].dt.day
df['Dmonth']=df['Delivery Date'].dt.month
df['Dyear']=df['Delivery Date'].dt.year
```

```python
df.drop(['Scheduled Date', 'Delivery Date'], axis = 1, inplace=True)
```

```python
pd.set_option('display.max_columns',None)
df.head(18)
```

```python
df.drop('Customer Location',axis=1,inplace=True)
```

```python
df['Transport'].value_counts()
```

```python
df['Material'].value_counts()
```

```python
df['Remote Location'].value_counts()
```

```python
df['Transport'] = df['Transport'].fillna("NaN")
df['Material'] = df['Material'].fillna("NaN")
df['Remote Location'] = df['Remote Location'].fillna("NaN")
```

```python
df.isnull().sum()
```

```python
df['Artist Reputation'] = df['Artist Reputation'].fillna(df['Artist Reputation'].mean())
df['Height'] = df['Height'].fillna(df['Height'].mean())
df['Weight'] = df['Weight'].fillna(df['Weight'].mean())
df['Width'] = df['Width'].fillna(df['Width'].mean())
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in df.columns:
  if df[i].dtype=='object':
    df[i]=le.fit_transform(df[i])

df.head()
```

```python
df['Cost']
```

```python
df['Cost'] = np.abs(df['Cost'].min())+1
df['Cost']=np.log(df['Cost'])
```

```python
x=df.drop('Cost',axis=1)
y=df['Cost']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
print(x_train.shape,y_train.shape)
x_test.shape,y_test.shape
```

```python
xcol=x_train.columns
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_train=pd.DataFrame(x_train,columns=xcol)
x_test=sc.transform(x_test)
x_test=pd.DataFrame(x_test,columns=xcol)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
lr=GradientBoostingRegressor()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_absolute_percentage_error
print(mean_absolute_error(y_test,pred))
print(np.sqrt(mean_squared_error(y_test,pred)))
mean_absolute_percentage_error(y_test,pred)
```
