```python
categorical_values=df.select_dtypes(include='object').columns
print('categorical_values:',categorical_values)
from sklearn.preprocessing import LabelEncoder
label_encoder={}
for columns in categorical_values:
    label_encoder[columns]=LabelEncoder()
    df[columns]=label_encoder[columns].fit_transform(df[columns])
```

```python
df1=df.copy()
df1.head()
```

```python
df1['Class'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
x=df1.drop('Class',axis=1)
y=df1['Class']
```

```python
smote=SMOTE()
x,y=smote.fit_resample(x,y)
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

# Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier

# K-Nearest Neighbors (KNN) Classifier
from sklearn.neighbors import KNeighborsClassifier

# AdaBoost Classifier
from sklearn.ensemble import AdaBoostClassifier

# XGBoost Classifier
from xgboost import XGBClassifier

# Support Vector Classifier (SVC)
from sklearn.svm import SVC

# Logistic Regression
from sklearn.linear_model import LogisticRegression

# Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB

# Polynomial Classifier (Pipeline for Polynomial Features + Logistic Regression)
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
```

```python
print(x.shape)
print(y.shape)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
print(y_test.value_counts())
y_train.value_counts()
```

```python
lr=LogisticRegression()
#y_train=y_train.to_numpy()
#y_train=y_train.reshape(-1)
#x_train=x_train.to_numpy()
#x_train=x_train.reshape(-1)
#smote=SMOTE()
#x_train,y_train=smote.fit_resample(x_train,y_train)
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```python
from sklearn.metrics import classification_report
models = {
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(n_estimators=100),
    'Gradient Boosting': GradientBoostingClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),
    'AdaBoost': AdaBoostClassifier(n_estimators=50),
    'XGBoost': XGBClassifier(),
    'Support Vector Classifier': SVC(kernel='rbf'),
    'Logistic Regression': LogisticRegression(max_iter=200),
    'Naive Bayes': GaussianNB(),
    'Polynomial + Logistic Regression': make_pipeline(PolynomialFeatures(degree=3), LogisticRegression(max_iter=200))
}
for model_name, model in models.items():
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    print('Result for'+ model_name+':')
    print(classification_report(y_test,predictions))
```
