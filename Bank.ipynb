```python
x=df.drop('Bankrupt?',axis=1)
y=df['Bankrupt?']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE

estimator = RandomForestClassifier(random_state=42)
rfe_selector = RFE(estimator=estimator, n_features_to_select=30, step=1)

rfe_selector = rfe_selector.fit(x, y)


selected_features = x.columns[rfe_selector.support_]
print("Selected Features (Wrapper Method):", list(selected_features))

x = x[selected_features]
```

```python
from imblearn.over_sampling import SMOTE
x,y=SMOTE().fit_resample(x,y)
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

# Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier

# K-Nearest Neighbors (KNN) Classifier
from sklearn.neighbors import KNeighborsClassifier

# AdaBoost Classifier
from sklearn.ensemble import AdaBoostClassifier

# XGBoost Classifier
from xgboost import XGBClassifier

# Support Vector Classifier (SVC)
from sklearn.svm import SVC

# Logistic Regression
from sklearn.linear_model import LogisticRegression

# Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB

# Polynomial Classifier (Pipeline for Polynomial Features + Logistic Regression)
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings('ignore')

```

```python
from sklearn.metrics import classification_report
models = {
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(n_estimators=100),
    'Gradient Boosting': GradientBoostingClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),
    'AdaBoost': AdaBoostClassifier(n_estimators=50),
    'XGBoost': XGBClassifier(),
    'Support Vector Classifier': SVC(kernel='rbf'),
    'Logistic Regression': LogisticRegression(max_iter=200),
    'Naive Bayes': GaussianNB(),
    #'Polynomial + Logistic Regression': make_pipeline(PolynomialFeatures(degree=3), LogisticRegression(max_iter=200))
}
for model_name, model in models.items():
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    print('Result for'+ model_name+':')
    print(classification_report(y_test,predictions))
```

```python
from sklearn.model_selection import RandomizedSearchCV
lorge=LogisticRegression()
param_rand={
    'C':[0.01,0.1,1],
    'penalty':['l1','l2'],
    'max_iter':[100,200],
    'solver':['liblinear','saga']
}
random_search=RandomizedSearchCV(estimator=lorge,param_distributions=param_rand,cv=3,scoring='accuracy',verbose=2)
random_search.fit(x,y)
print('Bestparameters:',random_search.best_params_)
print('Best Cross Validation Score:',random_search.best_score_)
```
