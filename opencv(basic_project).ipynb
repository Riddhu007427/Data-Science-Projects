```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Import the cv2_imshow function

# Step 1: Load the image
image = cv2.imread('/content/download.jpeg')  # Replace with your image path

# Step 2: Convert the image to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 3: Apply Gaussian Blur
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

# Step 4: Apply Canny Edge Detection
edges = cv2.Canny(gray_image, 100, 200)

# Step 5: Display all images using cv2_imshow for Colab
cv2_imshow(image)  # Show the original image
cv2_imshow(gray_image)  # Show the grayscale image
cv2_imshow(blurred_image)  # Show the blurred image
cv2_imshow(edges)  # Show the edge detection result

# Wait for a key press - not needed in Colab
# cv2.waitKey(0)
# cv2.destroyAllWindows()  # Not needed in Colab
```

```python
import cv2 # Face Detection with OpenCV
from google.colab.patches import cv2_imshow

# Load pre-trained Haar Cascade Classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Read an image (make sure the file path is correct)
image = cv2.imread('/content/download.jpeg')  # Replace with your image path

# Check if the image is loaded successfully
if image is None:
    print("Error: Image not loaded. Please check the file path.")
else:
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    # Draw rectangles around faces
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)  # Blue rectangle with thickness 2

    # Display the output image using cv2_imshow (in Colab)
    cv2_imshow(image)  # Only pass the image, no title
```

```python
import cv2 #Edge Detection Using Canny
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('/content/download.jpeg')  # Replace with your image path

# Perform some image processing (e.g., edge detection)
edges = cv2.Canny(image, 100, 200)

# Display the original image using cv2_imshow
cv2_imshow(image)  # Only pass the image, no title

# Display the processed image (edges) using cv2_imshow
cv2_imshow(edges)

# If you want to display with titles, use matplotlib (for both images):
# Original Image with Title
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')  # Hide axis
plt.show()

# Processed Image with Title
plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB), cmap='gray')
plt.title("Edges")
plt.axis('off')  # Hide axis
plt.show()
```

```python
import cv2  #Simple Image Transformation (Resize and Rotate)
from google.colab.patches import cv2_imshow

# Read the image
image = cv2.imread('/content/download.jpeg')  # Replace with your image path

# Check if the image was loaded successfully
if image is None:
    print("Error: Image not loaded. Please check the file path.")
else:
    # Resize the image (new width = 500, new height = 500)
    resized_image = cv2.resize(image, (500, 500))

    # Rotate the image (angle = 45 degrees)
    rows, cols, _ = image.shape
    M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 180, 1)
    rotated_image = cv2.warpAffine(image, M, (cols, rows))

    # Display the images using cv2_imshow (no titles here, only images)
    cv2_imshow(image)       # Display the original image
    cv2_imshow(resized_image)  # Display the resized image
    cv2_imshow(rotated_image)  # Display the rotated image
```
