```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('/content/fruits.jpeg')
cv2_imshow(img)
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
img=cv2.imread('/content/fruits.jpeg')
cv2_imshow(img)
img[:,:,0]=0  #color combination of GR   (BGR)
cv2_imshow(img)
```

```python
img=cv2.imread('/content/fruits.jpeg')
cv2_imshow(img)
img[:,:,1]=0  #color combination of BR   (BGR)
cv2_imshow(img)
```

```python
img=cv2.imread('/content/fruits.jpeg')
cv2_imshow(img)
img[:,:,2]=0  #color combination of BG   (BGR)
cv2_imshow(img)
```

```python
img=cv2.imread('/content/fruits.jpeg')
cv2_imshow(img)
img_GR=img[:,:,0]
img_BR=img[:,:,1]
img_BG=img[:,:,2]
new_img=np.hstack((img_GR,img_BR,img_BG))
cv2_imshow(new_img)
```

```python
#RESIZE
img_resize=cv2.resize(img,(500,500))
cv2_imshow(img_resize)
```

```python
img=cv2.imread('/content/fruits.jpeg')
img_resize=cv2.resize(img,(img.shape[1]//2,img.shape[0]//2))
cv2_imshow(img_resize)
```

```python
img_flip=cv2.flip(img,-1) #0-vertical flip,1-horizontal flip(rightside),-1 -vertical flip and horizontal flip
cv2_imshow(img_flip)
```

```python
#crop
img=cv2.imread('/content/fruits.jpeg')
img_crop=img[0:200,0:200]
cv2_imshow(img_crop)
```

```python
#save img
cv2.imwrite('img_crop.jpeg',img_crop)
```

```python
img1=cv2.imread('/content/img_crop.jpeg')
cv2_imshow(img1)
```

```python
#drawing rectangle
img2=np.zeros((500,500,3))
cv2_imshow(img2)
```

```python
img2=np.zeros((500,500,3))
cv2.rectangle(img2,pt1=(50,50),pt2=(200,200),color=(255,0,0),thickness=1)   #-1:to fill the square
cv2_imshow(img2)
```

```python
img2=np.zeros((500,500,3))
cv2.circle(img2,center=(250,250),radius=100,color=(0,0,255),thickness=5)
cv2_imshow(img2)
```

```python
img2=np.zeros((500,500,3))
cv2.line(img2,pt1=(250,250),pt2=(512,512),color=(0,255,0),thickness=5)
cv2_imshow(img2)
```

```python
img2=np.zeros((512,512,3))
cv2.putText(img2,org=(30,250),fontFace=0,color=(255,255,255),fontScale=2,text='Hello Ridesh!!!',thickness=2)
cv2_imshow(img2 )
```
